### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate 60-day rolling covariance of Algo 1 vs. S&P TSX 60 and plot the data
rolling_covariance = combined_returns['Algo 1'].rolling(window=60).cov(combined_returns['S&P TSX'])
rolling_covariance.plot(figsize=(20, 10), title='Rolling 60-Day Covariance of Algo 1 Returns vs. S&P TSX 60')

# Calculate 60-day rolling variance of Algo 1 vs. S&P TSX 60and plot the data
rolling_variance = combined_returns['Algo 1'].rolling(window=60).var()
rolling_variance.plot(figsize=(20, 10), title='Rolling 60-Day Variance of S&P 500 Returns')

# Calculate 60-day rolling beta of Algo 1 and plot the data
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of Algo 1')

plt.title("Algo 1 and S&P TSX 60")
plt.xlabel("Date")
plt.ylabel("Beta")
plt.legend(["Algo 1"])
plt.show()
           
           
    
```

## Rolling Statistics Challenge: Exponentially Weighted Average 

An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half life for each portfolio, using standard deviation (`std`) as the metric of interest.

```python
# Use `ewm` to calculate the rolling window
ewma_21 = combined_returns.ewm(halflife=21).mean()

# Plot the EWMA
ewma_21.plot(figsize=(12, 8))
plt.title("Exponentially Weighted Moving Average (21-Day) of Portfolio Returns")
plt.xlabel("Date")
plt.ylabel("EWMA")
plt.legend(loc='best')
plt.show()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
# Calculate the annualized average return
annualized_mean = combined_returns.mean() * 252

# Calculate the annualized standard deviation (assuming 252 trading days in a year)
annualized_std = combined_returns.std() * (252 ** 0.5)

# Calculate the Sharpe ratios (assuming a risk-free rate of 0%)
sharpe_ratios = annualized_mean / annualized_std
```

```python
# Visualize the sharpe ratios as a bar plot
# Plot the Sharpe ratios
sharpe_ratios.plot.bar(figsize=(10, 6))
plt.title("Sharpe Ratios of Portfolios")
plt.ylabel("Sharpe Ratio")
plt.show()
```

```python
# Compare the Sharpe ratios of algorithmic strategies to the market and whales portfolios
algo_sharpe_ratios = sharpe_ratios[['Algo 1', 'Algo 2']]
other_sharpe_ratios = sharpe_ratios.drop(['Algo 1', 'Algo 2'])

outperformance = algo_sharpe_ratios > other_sharpe_ratios.max()
print("Do algorithmic strategies outperform the market and whales portfolios?")
print(outperformance)
```
